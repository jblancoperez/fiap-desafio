```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
arquivo = pd.read_csv('c:/Users/Angelo/Downloads/solicitacoescredito.csv')
arquivo.info()
```

```python
#Remover transacaoes que nao passaram por analise de credito
arquivo['percentualProtestos'].fillna(0,inplace=True)
limpo = arquivo.drop(['numero_solicitacao','razaoSocial','nomeFantasia','cnpjSemTraco','ativoCirculante','passivoCirculante','totalAtivo','totalPatrimonioLiquido','endividamento','duplicatasAReceber','estoque','anoFundacao','intervaloFundacao','anoFundacao','capitalSocial','periodoDemonstrativoEmMeses','periodoBalanco','custos'], 1)
limpo.info()
```

```python
statusvalidos = ['EmAnaliseDocumentacao','AguardandoAprovacao','DocumentacaoReprovada']
filtrado = limpo.loc[~limpo['status'].isin(statusvalidos)]
filtrado.head(50)
#filtrado.info()
```

```python
filtrado.isnull().sum()
```

```python
filtrado.head()
```

```python
filtrado.definicaoRisco.unique()
```

```python
#Converter def Risco
defRisco = pd.get_dummies(filtrado['definicaoRisco'],drop_first=False)
filtrado = pd.concat([filtrado.drop(['definicaoRisco'],1),defRisco],1)
filtrado
```

```python
#Conveter status
status = pd.get_dummies(filtrado['status'],drop_first=False)
filtrado = pd.concat([filtrado.drop(['status'],1),status],1)
filtrado
```

```python
fig = plt.subplots(figsize = (15,9))
fig = sb.heatmap(filtrado.corr())  
fig
```
